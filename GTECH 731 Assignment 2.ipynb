```python
with open(r"C:\Users\alexf\OneDrive\Desktop\Geocomp\Data\gz_2010_us_050_00_20m.json") as f:
    data = json.load(f)
```

```python
with open(r"C:\Users\alexf\OneDrive\Desktop\Geocomp\Data\fipsToState.json") as sList:
    stateDict = json.load(sList)
```

```python
# Task 2: Find the three most common county names in the USA

# the keys in data are ['type','features']

# The value of the key 'features' is a list with thousands of elements
# Each element of that list is a dictionary
# Eor example, data['features'][n] is a dictionary with the 'features' of the nth county
#
# The keys of the dictionary data['features'][n] are ['type','properties','geometry']

# The value of the key 'properties' in the dictionary data['features'][n] is another dictionary
# with the keys ['GEO_ID','STATE','COUNTY','NAME','LSAD','CENSUSAREA']

# In this Task, we will be interested in the county names, accessed as the values of
# data['features'][n]['NAME'] for county n

# The value of the key 'geometry' in the dictionary data['features'][n] is a dictionary
# with the keys ['type','coordinates']
# We don't use that data in this project, but it would be cool to try to use that data
# to draw the counties using some graphical output.

featureList = data['features']
# featureList is a list with len(data['features']) 'features'
# The value of each 'feature' is a dictionary with keys ['type','properties','geometry']
# The value of the key 'properties' is a dictionary with the keys ['GEO_ID','STATE','COUNTY','NAME','LSAD','CENSUSAREA']
# The name of the nth county is the value of the key featureList[n]['properties']['NAME']
# If f is an element of featureList[n], then f['properties']['NAME'] is the name of the nth county

cntyDict = {}
# This creates a blank dictionary that we will fill with keys consisting of county names with values
# consisting of the number of counties with that name

for f in featureList:
    # Here we step through each 'feature' in featureList, each corresponding to a county in the USA.

    name = f['properties']['NAME']
    # name is the name of the county in question
    theState = f['properties']['STATE']
    # theState is the name of the state that includes the county in question

    if cntyDict.get(name) is None:
        cntyDict.update({name: 1})
        # If this county's name is not yet a key in the dictionary cntyDict, then add this name as a new key
        # and give that key the value 1, indicating that there is only one county (so far) with that name.

    else:
        cntyDict.update({name: cntyDict.get(name) + 1})
        # increment the value of the key name by one
        # i.e. add one more county with this name

cntyDictSorted = sorted(cntyDict.items(), key=lambda item: item[1], reverse=True)
# dict.items() is a list of the key:value pairs in the dictionary dict
# We use the method items() here in the function sorted() which sorts lists.
# In the method sorted, the argument key=lambda indicates what aspect of the list we are sorting on

print()
print('The three most common county names in the United States are ', end="")
print(cntyDictSorted[0][0], '\b, ', end="")
print(cntyDictSorted[1][0], '\b, and ', end="")
print(cntyDictSorted[2][0], '\b.')
print()

for n in range(0, 3):
    print('There are', cntyDictSorted[n][1], 'counties named', cntyDictSorted[n][0], end="")
    print(', one in each of the following states:')

    for f in featureList:

        if f['properties']['NAME'] == cntyDictSorted[n][0]:
            print("    ", end="")
            print(stateDict.get(f['properties']['STATE']))

    print()
```

```python
# Task 3: Basic statistics by state

stateCntys = {}
# stateCntys is a dictionary that will be filled with keys:values corresponding to 'state name':[list of counties]

for f in featureList:

    county = f['properties']['NAME']
    state = stateDict.get(f['properties']['STATE'])
    size = f['properties']['CENSUSAREA']

    # create the dictionary stateCntys:
    if stateCntys.get(state) is None:
        stateCntys.update({state: [[county, size]]})
    else:
        countyList = stateCntys.get(state)
        countyList.append([county, size])
        stateCntys.update({state: countyList})

stateList = list(stateCntys)

for s in stateList:
    numCounties = len(stateCntys.get(s))

    countySizes = stateCntys.get(s)
    countySizesSorted = sorted(countySizes, key=lambda blah: blah[1], reverse=True)

    roundVal = 1
    # This specifies the number of decimal places to exhibit in the outputted areas

    # compute total area of counties in the state
    totalArea = 0
    for cNum in range(numCounties):
        totalArea = totalArea + countySizes[cNum][1]
    totalArea = round(totalArea, roundVal)
    aveArea = round(totalArea / numCounties, roundVal)

    # print out data about the counties in each state
    if numCounties == 1:
        print(s, 'has only one county.')
        print('    ', end="")
        print(countySizesSorted[0][0], 'has a land area of', round(countySizesSorted[0][1], roundVal), 'square miles.')
    else:
        print(s, 'has', numCounties, 'counties.')
        print('    ', end="")
        print('The largest county in', s, 'is', countySizesSorted[0][0], 'with a land area of',
              round(countySizesSorted[0][1], roundVal),
              'square miles.')
        print('    ', end="")
        print('The smallest county in', s, 'is', countySizesSorted[numCounties - 1][0], 'with a land area of',
              round(countySizesSorted[numCounties - 1][1], roundVal), 'square miles.')
        print('    ', end="")
        print('The total land area of all of the counties in', s, 'is', totalArea, 'square miles.')
        print('    ', end="")
        print('The average land area of the counties in', s, 'is', aveArea, 'square miles.')
    print()
```
